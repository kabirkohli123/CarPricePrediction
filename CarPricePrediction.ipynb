```python
from sklearn.preprocessing import StandardScaler
scalar = StandardScaler() 
```

```python
df['Kms_driven']
```

```python
Y = df['Price']
Y
```

```python
X = df.drop(['Price'],axis='columns')
X
```

```python
Year = np.array(X['Year'])
Year = Year.reshape(-1,1)
Year
```

```python
X['Year'] = scalar.fit_transform(Year)
X['Year']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,train_size=0.8,random_state=1)
```

```python
X_train
```

```python
X_test
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train,Y_train)
```

```python
Y_predict = lr.predict(X_test)
print("The first 10 predicted values of Y is:")
print(Y_predict[:10])
print("The first 10 test values of Y is:")
print(np.array(Y_test[:10]))
```

```python
from sklearn.metrics import r2_score
r2 = r2_score(Y_test,Y_predict)
r2
```

```python
lr.score(X_test,Y_test)
```

```python
from sklearn.model_selection import cross_val_score
print(cross_val_score(lr,X,Y,cv=10))
print(np.average(cross_val_score(lr,X,Y,cv=10)))
```

```python
from sklearn.linear_model import Lasso
lasso = Lasso()
lasso.fit(X_train,Y_train)
```

```python
Y_predict = lasso.predict(X_test)
print("The first 10 predicted values of Y is:")
print(Y_predict[:10])
print("The first 10 test values of Y is:")
print(np.array(Y_test[:10]))
```

```python
lasso.score(X_test,Y_test)
```

```python
print(cross_val_score(lasso,X,Y,cv=10))
print(np.average(cross_val_score(lasso,X,Y,cv=10)))
```

```python
from sklearn.linear_model import Ridge
ridge = Ridge()
ridge.fit(X_train,Y_train)
```

```python
Y_predict = ridge.predict(X_test)
print("The first 10 predicted values of Y is:")
print(Y_predict[:10])
print("The first 10 test values of Y is:")
print(np.array(Y_test[:10]))
```

```python
ridge.score(X_test,Y_test)
```

```python
r3 = r2_score(Y_test,Y_predict)
r3
```

```python
print(cross_val_score(ridge,X,Y,cv=10))
print(np.average(cross_val_score(ridge,X,Y,cv=10)))
```

```python
from sklearn.tree import DecisionTreeRegressor
dtr = DecisionTreeRegressor()
dtr.fit(X_train,Y_train)
Y_predict = dtr.predict(X_test)
print("The first 10 test value of Y is ")
print(np.array(Y_test[:10]))
print("The first 10 predicted value of Y is ")
print(Y_predict[:10])
print("The accuracy of Decision Tree is ")
print(dtr.score(X_test,Y_test))
```

```python
print(cross_val_score(dtr,X,Y,cv=10))
print(np.average(cross_val_score(dtr,X,Y,cv=10)))
```
